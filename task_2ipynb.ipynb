```python
import pandas as pd
```

```python
# 2. Load the dataset directly from the internet link
file_url = "https://raw.githubusercontent.com/yannie28/Global-Superstore/master/Global_Superstore(CSV).csv"
df = pd.read_csv(file_url, encoding='latin1')

# 3. Look at the first 5 rows to make sure it loaded correctly
df.head()
```

### Data Inspection: Structural Health

```python
# 1. Handle missing values by dropping rows with missing essential data
df_cleaned = df.dropna()

# 2. Drop duplicate rows
df_cleaned = df_cleaned.drop_duplicates()

# 3. Convert 'Order Date' column to a proper date format
# (If your dataset uses 'Ship Date', you can convert that too!)
df_cleaned['Order Date'] = pd.to_datetime(df_cleaned['Order Date'])

# 4. Verify the changes worked
print("New Data Shape (Rows, Columns):", df_cleaned.shape)
print("\nNew Data Types:")
print(df_cleaned[['Order Date']].dtypes)
```
